### Stage 2 — Make it a record, add a function and a decision

**Lesson.** A **dictionary** groups related fields under labels, much tidier than five loose
variables. A **function** names a piece of reusable work. **if/elif/else** lets your code
decide something.

**Task.**
1. Put the book into a dict: `book = {"title": ..., "pages": ..., "minutes": ..., "genre": ...}`.
2. Write `pace(book)` that returns pages per minute (`pages / minutes`).
3. Write `speed_label(book)` that returns `"fast"`, `"steady"` or `"slow"` using if/elif/else
   on the pace (suggested: > 0.8 fast, > 0.6 steady, else slow).
4. Print the label for your book.

> 💾 Commit: `git commit -m "Stage 2: book record, pace() and speed_label()"`

```python
# 1. Put the book into a dictionary

book = {
    'Title': 'Alice in Wonderland',
    'Pages': 500,
    'Minutes': 2000,
    'Genre': 'fantasy',
    'Finished': False
}
print(book)
```

```python
# 2. Write `pace(book)` that returns pages per minute (`pages / minutes`)

def pace(book):
    return book['Pages'] / book['Minutes']

print(pace(book))
```

```python
# 3. Write `speed_label(book)` that returns `"fast"`, `"steady"` or `"slow"` using if/elif/else
#    on the pace (suggested: > 0.8 fast, > 0.6 steady, else slow).

def speed_label(book):
    reading_pace = pace(book)

    if reading_pace > 0.8:
        return "fast"
    elif reading_pace > 0.6:
        return "steady"
    else:
        return "slow"
```

```python
# Print the label for the book

print(speed_label(book))
```

Stage 3 — Many books (a list) and a loop
Lesson. One book is no fun. A list holds many records, a for loop walks through them, and len() counts them.

A starter dataset is in the next cell, run it so everyone's working with the same books.

```python
# (given) starter data — run this cell

books = [
    {"title": "Dune",                 "pages": 412, "minutes": 540, "genre": "sci-fi"},
    {"title": "The Hobbit",           "pages": 310, "minutes": 360, "genre": "fantasy"},
    {"title": "Sapiens",              "pages": 498, "minutes": 720, "genre": "non-fiction"},
    {"title": "Project Hail Mary",    "pages": 476, "minutes": 600, "genre": "sci-fi"},
    {"title": "The Name of the Wind", "pages": 662, "minutes": 840, "genre": "fantasy"},
    {"title": "Educated",             "pages": 334, "minutes": 480, "genre": "non-fiction"},
]
```

Task. Loop over books and print each book's title and page count. Then add up the total pages across all books using a loop, and print it. Also print how many books there are.

```python
# Loop through each book in the list
total_pages = 0

for book in books:
    # Print the current book's title and pages
    print(f"Title: {book['title']}, Pages: {book['pages']}")

    # Add the current book's pages to the running total
    total_pages +=book['pages']

print(f"Total Pages: {total_pages}")  # Print the final calculated page total
print(f"Total Books: {len(books)}")  # Print the total number of books in the list
```

Stage 4 — Group by genre (dicts + nesting)
Lesson. A list of dicts is your first taste of nesting, a structure inside a structure. It's exactly the shape real datasets arrive in. A common move is to walk the nested data and build a new dict that summarises it.

Task. Build pages_by_genre, a dict mapping each genre to its total pages, e.g. {"sci-fi": 888, "fantasy": 972, ...}. The dict method .get(genre, 0) is the clean way to start each running total at zero. Then loop over your result and print each genre with its total.

Bonus: also build count_by_genre (how many books per genre).

💾 Commit: git commit -m "Stage 4: group by genre"

```python
# Empty dictionaries
pages_by_genre = {}
book_counts = {}

# Loop for looking up books in books
for book in books:
    genre = book["genre"]  # Extract the book genre
    pages = book["pages"]  # Extract the book count

    ## Add pages to genre total (starts at 0 if genre is new)
    pages_by_genre[genre] = pages_by_genre.get(genre, 0) + pages

# Add 1 to genre book count (starts at 0 if genre is new)
    book_counts[genre] = book_counts.get(genre, 0) + 1

# Loop through page totals and print them    
for genre, total_pages in pages_by_genre.items():
    print(f"{genre}: {total_pages} pages")

# Loop through book counts and print them
for genre, count in book_counts.items():
    print(f"{genre}: {count} book(s)")
```
